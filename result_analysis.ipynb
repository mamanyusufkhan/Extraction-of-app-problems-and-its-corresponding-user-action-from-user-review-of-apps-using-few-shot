```python
def construct_accuracy_precision_and_recall(tn, fp, fn, tp):
  accuracy = (tp + tn)/(tp + fp + fn + tn)
  print("Accuracy: " + str(accuracy))


  if tp == 0 and fp == 0:
    precision = 0
  else:
    precision = tp / (tp + fp)
  print("Precision: "+ str(precision))


  recall = tp / (tp + fn)
  print("Recall: "+ str(recall))
```

```python
print("Overall:")
construct_accuracy_precision_and_recall(130, 15, 8, 47)
```

```python
print("Keywords:")
construct_accuracy_precision_and_recall(6, 2, 2, 13)
```

```python
print("Without Keywords: ")
construct_accuracy_precision_and_recall(124, 13, 6, 34)
```
